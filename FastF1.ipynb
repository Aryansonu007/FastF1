```python
year=2025
circuit="Austria"
event="R"
Aus=fastf1.get_session(year,circuit,event)
Aus.load()
```

```python
Aus.results
```

```python
Aus.results.columns
```

```python
Aus.results[['DriverId','Time']]
```

```python
Aus.results.Time.dtype           #Strange Datatype aint it > It is TimedeltaArray
```

Lets try to convert this to float

```python
Aus.results.cstime=Aus.results.Time.to_string()
```

```python
Aus.results.ctime=[i for i in Aus.results.ctime][-9:-1]
```

```python
Aus.results.ctime
```

```python
Aus.results.cstime
```

```python
v1=lambda x: x.total_seconds()
Aus.results.ctime=Aus.results.Time.apply(v1)
```

```python
Aus.results.ctime
```

```python
for index, time_value in Aus.results.ctime.items():
  if time_value == 5027.693:
    continue
  else:
    Aus.results.ctime[index] = time_value + 5027.693
```

```python
Aus.results.ctime
```

```python
plt.figure(figsize=(20,10))
plt.plot(Aus.results.DriverId,Aus.results.ctime,marker='.')
```

```python
Max=abd.laps.pick_drivers("VER")
```

```python
Max
```

```python
Max.describe()
```

```python
lap=58
lap_specific=Max[(Max["LapNumber"]==lap)]
lap_specific
```

```python
lap_specific.columns
```

```python
lap_data=lap_specific.get_car_data()
```

```python
lap_data
```

```python
from fastf1 import plotting
plotting.setup_mpl()
fig,ax=plt.subplots(figsize=(20,10))
ax.plot(lap_data.Time,lap_data.Speed,label="BHRUUUM",color='red')
ax.set_xlabel("TIME")
ax.set_ylabel("SPEED")
ax.set_title("Speed Vs Time")
ax.grid()
ax.set_facecolor("black")

ax.legend()
```

```python
Ham=abd.laps.pick_driver("HAM")
Ham
```

```python
lap=58
lap_specific=Ham[(Ham["LapNumber"]==lap)]
lap_specific
```

```python
hl=lap_specific.get_car_data()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(lap_data.Time, lap_data.Speed, label='VER Speed', color='red')
ax.plot(hl.Time, hl.Speed, label='HAM Speed', color='cyan')

ax.set_xlabel("Time")
ax.set_ylabel("Speed")
ax.set_title("Speed Comparison: VER vs HAM - Lap 58")
ax.grid()
ax.legend()
ax.set_facecolor("black")
plt.show()
```
